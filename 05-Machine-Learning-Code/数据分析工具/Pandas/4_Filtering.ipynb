```python
people = {
    "first": ["Corey", 'Jane', 'John'], 
    "last": ["Schafer", 'Doe', 'Doe'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df['last'] == 'Doe'
```

```python
filt = (df['last'] == 'Doe')
filt
```

```python
df[filt]
```

```python
df.loc[filt]
```

```python
df.loc[filt,'email']
```

```python
filt = (df['last'] == 'Doe') & (df['first'] == 'John')
```

```python
df.loc[filt,'email']
```

```python
filt = (df['last'] == 'Schafer') | (df['first'] == 'John')
```

```python
df.loc[filt,'email']
```

```python
df.loc[~filt,'email']
```
