```python
import matplotlib.pyplot as plt
import csv

x = []
y = []

with open('example.txt','r',encoding='utf-8') as csvfile:
    plots = csv.reader(csvfile,delimiter='，')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))

        
plt.plot(x,y,label='load from file')
# plt.xlabel('x')
# plt.ylabel('y')
plt.title('Interesting Graph')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 

x,y = np.loadtxt('example.txt',delimiter='，',encoding='utf-8',unpack=True)
        
plt.plot(x,y,label='load from file')
# plt.xlabel('x')
# plt.ylabel('y')
plt.title('Interesting Graph')
plt.legend()
plt.show()
```
