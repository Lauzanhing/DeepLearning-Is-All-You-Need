```python
import requests
import re
```

```python
r = requests.get("https://editorialexpress.com/conference/AFA2020/program/AFA2020.html")
#print(r.text)
pattern = re.compile(r'paper_id=(\d+)')
id_list = re.findall(pattern,r.text)
```

```python
print(len(id_list))
print(id_list)
```

```python
pattern2 = re.compile(r'paper_id=.*">(.*)</a></td><')
name_list = re.findall(pattern2,r.text)
#print(len(name_list))
print(name_list)
```

```python
for i in (0,1):
    filename = name_list[i] + '.pdf'
    new_url = "https://editorialexpress.com/cgi-bin/conference/download.cgi?db_name=AFA2020&paper_id="+str(id_list[i])
    print(new_url)
    down_url = requests.get(new_url)
    print(down_url)
    fo = open(filename,'wb')
    fo.write(down_url.content)
    print(filename+"下载完成")
    fo.close
```
